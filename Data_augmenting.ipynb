Generating Text based on given Class

```python
def generate_text(prompt, class_label):

    model_engine = "gpt-3.5-turbo"  
    # model_engine = "gpt-3.5-turbo-0301"

    model_prompt = (f"Considering the given topic, generate similar text to the given text.\nTopic: {class_label}\nText: {prompt}####\n\nSimilar text:")  # Set up the GPT prompt

    time.sleep(0.01)
    response = openai.ChatCompletion.create(
        model=model_engine,
        messages=[{"role": "user", "content": model_prompt}],
        temperature = 0.5,
        stop="####"
    )

    print(prompt)
    print(response['choices'][0]['message']['content'])
    
    return response['choices'][0]['message']['content']
```

Generating Explanation based on given augmented text

```python
def generate_explanation(ori_text,ori_exp,aug_text):
        
    model_engine = "gpt-3.5-turbo" 
 
    model_prompt = (f"Consider the examples and generate a very short explanation of the given text.\n\n \
    text: i want to be dead, everyone hates me\nexplanation: everyone hates me\n \
    text: If reincarnation is a real thing I was a selfish person that wanted the world now im stuck\nexplanation: selfish person \n \
    text: I can't even talk to anyone anymore because everyone walks around with those goddamn airpods and it makes everyone so unapproachable\nexplanation: makes everyone so unapproachabley\n \
    text: I have no motivation nor intention to progress in my life\nexplanation: no motivation \n \
    text: What if my parents die? What do I do? I don't know, I feel hideously ugly\nexplanation: feel hideously ugly\n \
    text: {ori_text}\nexplanation: {ori_exp}\n\n \
    text: {aug_text}\nexplanation:")  
    # print(model_prompt)

    time.sleep(0.01)
    response = openai.ChatCompletion.create(
        model=model_engine,
        messages=[{"role": "user", "content": model_prompt}],
        temperature=0.2
        # presence_penalty=-1.5
    )
    print(ori_exp)
    print(response['choices'][0]['message']['content'])
    print("\n\n")
    return response['choices'][0]['message']['content']
```

```python
# Declare global lists to store results

augmented_data1 =[]
augmented_data2 =[]
augmented_data_labels =[]
original_data = []
original_explain = []
```

```python
def appending_data(text, label, explanation, original, ori_exp):

  augmented_data1.append(text)
  augmented_data_labels.append(label)
  augmented_data2.append(explanation)
  original_data.append(original) 
  original_explain.append(ori_exp) 
```

```python
# Call API to generate text

class_1=0
class_2=0
class_3=0
class_4=0
max_records=1139
i=0

# df = df.reset_index()  # make sure indexes pair with number of rows
for index, row in df.iterrows():

  original_text = row['Text']
  original_exp = row['Explanations']
  class_label = row['Aspect']

  # original_token_ratio = calc_token_ratio(original_text, original_exp)

  print(i)
  i+=1

  if(class_label==1 and c1+class_1 < max_records):
    augmented_text = generate_text(original_text, "Physical Aspect")
    augmented_exp = generate_explanation(original_text, original_exp, augmented_text)
    appending_data(augmented_text, class_label, augmented_exp, original_text, original_exp)
    class_1+=1
    
  elif (class_label==2 and c2+class_2 < max_records):
    augmented_text = generate_text(original_text, "Intellectual and Vocational Aspect")
    augmented_exp = generate_explanation(original_text, original_exp, augmented_text)
    appending_data(augmented_text, class_label, augmented_exp, original_text, original_exp)
    class_2+=1

  elif (class_label==3 and c3+class_3 < max_records): 
    augmented_text = generate_text(original_text, "Social Aspect")
    augmented_exp = generate_explanation(original_text, original_exp, augmented_text)
    appending_data(augmented_text, class_label, augmented_exp, original_text, original_exp)
    class_3+=1

  elif (class_label==4 and c4+class_4 < max_records):
    augmented_text = generate_text(original_text, "Spiritual and Emotional Aspect")
    augmented_exp = generate_explanation(original_text, original_exp, augmented_text)
    appending_data(augmented_text, class_label, augmented_exp, original_text, original_exp)
    class_4+=1
```

```python
New_dataset = pd.DataFrame({'Original_text': original_data, 'Augmented_Text': augmented_data1, 'Aspect': augmented_data_labels, 'Original_Explanation': original_explain, 'Augmented_Explanation': augmented_data2})
```
