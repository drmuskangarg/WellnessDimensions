seperating original samples which were not used for augmenting

```python
def clear_testSet(original, augmented):
  # Get records from original not included in augmented dataset based on 'Text' column
  pure_original = original[~original['Text'].isin(augmented['Original_text'])]

# ----------------------------------------------
  # # Create an empty dataframe to store the samples
  testset = pd.DataFrame(columns=pure_original.columns)
  pure_original_ununsed = pd.DataFrame(columns=pure_original.columns)

  # Loop through unique classes in the dataframe
  for class_label in pure_original['Aspect'].unique():
      # Filter the dataframe by class
      class_df = pure_original[pure_original['Aspect'] == class_label]
      # Take first 45 samples from the filtered dataframe
      samples = class_df.head(n=45)
      # Append the samples to the samples_df
      testset = testset.append(samples)
      # Remove the samples from the original dataframe

  pure_original_ununsed = pure_original.drop(testset.index)
# ----------------------------------------------

  testset.to_csv("Stage2/Training_Testing_Data/Aspect_testingset.csv", index=False)
  print(testset['Aspect'].value_counts())
  # -------------------------------------------------

  # Get records from original which used in augmented dataset based on 'Text' column
  used_original = original[original['Text'].isin(augmented['Original_text'])]

  # now combine used originals with additional unused originals
  training_original = pd.concat([pure_original_ununsed, used_original], axis=0)

  training_original.to_csv("Stage2/Training_Testing_Data/s1_original_dataset.csv", index=False)
  print(training_original['Aspect'].value_counts())
```

```python
clear_testSet(original, d1)
```
