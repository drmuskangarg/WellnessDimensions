Plotting average semantic similarities of "Text"

```python
# function to add value labels
# def addlabels(x,y):
#     for i in range(len(x)):
#         plt.text(i,y[i],"{:.2f}".format(y[i]),size="small", weight="bold", horizontalalignment="center", verticalalignment="baseline", ma="center")

def addlabels(x,y):
  for x,y in zip(x,y):

      label = "{:.2f}".format(y)

      plt.annotate(label, # this is the text
                  (x,y), # these are the coordinates to position the label
                  textcoords="offset points", # how to position the text
                  xytext=(0,0), # distance from text to points (x,y)
                  ha='center', # horizontal alignment can be left, right or center
                  va='bottom')

def addlabels2(x,y):
  for x,y in zip(x,y):

      label = "{:.2f}".format(y)

      plt.annotate(label, 
                  (x,y), 
                  textcoords="offset points", 
                  xytext=(10,10), 
                  ha='left',
                  va='center')
      
def addlabels3(x,y):
  for x,y in zip(x,y):

      label = "{:.2f}".format(y)

      plt.annotate(label, 
                  (x,y), 
                  textcoords="offset points", 
                  xytext=(20,20), 
                  ha='right',
                  va='top')
```

```python
x = similarity_model_names

y1 = turbo1_sim_text
y2 = turbo2_sim_text
y3 = curie_sim_text
y4 = davinci_sim_text
y5 = eda_sim_text
y6 = bt_sim_text

fig, ax = plt.subplots(figsize=(10,8))

plt.plot(x, y1, label='gpt-3.5-turbo', marker='o')
plt.plot(x, y2, label='gpt-3.5-turbo-0301', marker='o')
plt.plot(x, y3, label='text-curie-001', marker='o')
plt.plot(x, y4, label='text-davinci-003', marker='o')
plt.plot(x, y5, label='EDA', marker='o')
plt.plot(x, y6, label='BT', marker='o')


plt.legend(loc="lower right")
plt.grid()
plt.xticks(rotation=45, ha='right',fontsize=12)

addlabels3(x, y1)
addlabels(x, y2)
addlabels3(x, y3)
addlabels(x, y4)
addlabels(x, y5)
addlabels(x, y6)


#add horizontal line at mean value of y
plt.axhline(y=np.nanmean(y1), color='blue', linestyle='--', linewidth=3, label='Avg')
plt.axhline(y=np.nanmean(y2), color='orange', linestyle='--', linewidth=3, label='Avg')
plt.axhline(y=np.nanmean(y3), color='green', linestyle='--', linewidth=3, label='Avg')
plt.axhline(y=np.nanmean(y4), color='red', linestyle='--', linewidth=3, label='Avg')
plt.axhline(y=np.nanmean(y5), color='purple', linestyle='--', linewidth=3, label='Avg')
plt.axhline(y=np.nanmean(y6), color='brown', linestyle='--', linewidth=3, label='Avg')

plt.xlabel("Similarity Models", fontsize=12)
plt.ylabel("Average Semantic/Syntactic Similarity", fontsize=12)
# plt.title('Average Textual Similarity among Original and Augmented Text', fontsize=13)

plt.savefig('Stage2/Results/S1_Text_average_similarity.pdf', bbox_inches="tight", pad_inches=0.1)  
plt.show()
```
