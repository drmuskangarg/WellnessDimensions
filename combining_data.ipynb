```python
def create_finalDataset(original, augmented, filename):
  original = original.drop('Unnamed: 0', axis=1)
  augmented = augmented.drop(['Original_text', 'Original_Explanation'], axis=1)

  frames = [original, augmented.rename(columns={'Augmented_Text':'Text', 'Augmented_Explanation':'Explanations'})]
  final = pd.concat(frames, ignore_index=True)
  print(final['Aspect'].value_counts())

  final.to_csv("Training_Testing_Data/s1_training_dataset_"+filename+".csv", index=False)
```

```python
create_finalDataset(original, augmented1, "chatgpt1_1_0")
```
